```python
### DRIVE DATA ###
from google.colab import drive

drive.mount('/content/drive', force_remount=True)
PATH = "/content/drive/My Drive/TFM/Data/FakeNews_Task3_2022/"
```

# **ESTUDIO TRAINING**

```python
# Análisis de Distribución #
df_train = pd.read_csv(PATH + "Task3_train_dev/Task3_english_training.csv") 
df_train = df_train[['text', 'our rating']]
df_train = df_train.rename(columns = {"text": "Text", "our rating": "Label"})
print('Conjunto de Entrenamiento: ', df_train.shape)

counts = df_train['Label'].value_counts()
print(counts)

counts_train = df_train['Label'].value_counts().to_dict()

fig = plt.figure(figsize = (10, 6))
LABELS = counts_train.keys()
values = counts_train.values()

plt.bar(LABELS, values, width = 0.4)
plt.xlabel("Clases")
plt.ylabel("Número de Textos")
plt.title("Distribución de las clases en el Conjunto de Entrenamiento")
plt.show()
```

```python
# Análisis de Longitud #
df_train['Length'] = df_train['Text'].apply(lambda text: len(text.split()))
df_train['Length'].describe()
```

```python
plt.hist(df_train['Length'], color = 'blue', edgecolor = 'black', bins = int(60/5))
plt.title('Distribución de la Longitud de los Textos')
plt.xlabel('Tamaño (Tokens)')
plt.ylabel('Número de textos')
```

```python
# Densidad de Clases #
sns.kdeplot(data=df_train, x='Length', hue='Label')

plt.legend(LABELS)
plt.title('Densidad con Multiples Clases')
plt.xlabel('Tamaño')
plt.ylabel('Densidad')
```
